```python
cap=cv2.VideoCapture("KneeBendVideo.mp4")
#setup the mediapipe model
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose:
    while True:
    #capture.read reads the video frame by frame
         isTrue,frame=cap.read()
    
    #Recolouring the image to rgb,frame is the snapshot 
         image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
         image.flags.writeable= False
    
    #accessing pose model as storing result in results
         results=pose.process(image)
    
    #recolouring the image to bgr
         image.flags.writeable= True
         image=  cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
         
    #render detection drawspec is the specification on how we want to draw ,the next drawig spec is connection drawing spec,first joint second skeleton(connections)
         mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,mp_drawing.DrawingSpec(color=(235,107,64),thickness=2,circle_radius=2),mp_drawing.DrawingSpec(color=(245,117,66),thickness=2, circle_radius=2))
    
         cv2.imshow("Mediapipe Feed",image)
         if cv2.waitKey(10)& 0xFF== ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

# Calculate Angles

```python
def calculate_angle(a,b,c):
    a=np.array(a) #first point(24)
    b=np.array(b) #mid
    c=np.array(c) #last ankle point
    
    radians=np.arctan2(c[1]-b[1],c[0]-b[0])-np.arctan2(a[1]-b[1],a[0]-b[0]) #a[0] means first point x
    angle=np.abs(radians*180/np.pi)
    
    if angle>180.0:
        angle= 360-angle
    return angle
```
